```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
folder_path = '../data/IEEE-CIS-Fraud-Detection/'
```

```python
df_xgbm = pd.read_pickle(f'{folder_path}/oof_stacking_xgbm_tss5', compression='gzip').rename(columns={'isFraud':'xgbm_isFraud'})
```

```python
df_lgbm = pd.read_pickle(f'{folder_path}/oof_stacking_lgbm_tss5', compression='gzip').rename(columns={'isFraud':'lgbm_isFraud'})
```

```python
sub = pd.merge(df_xgbm, df_lgbm, how='left', on='TransactionID')
sub['isFraud'] = sub.drop(columns=['TransactionID']).mean(axis=1)
```

```python
sub
```

```python
sub[['TransactionID', 'isFraud']].to_csv(f'{folder_path}/blending_lgbm_xgbm_tss5.csv',index=False)
```
